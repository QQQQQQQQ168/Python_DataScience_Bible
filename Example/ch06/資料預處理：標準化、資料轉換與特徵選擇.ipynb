# 數值資料標準化

## Z 分數標準化：StandardScaler

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('汽車車型資料檔.csv')
std = StandardScaler()
df1 = df[['哩/加侖','氣缸','排氣量','馬力','重量','加速時間','年份']]
arr = std.fit_transform(df1)
df[['哩/加侖','氣缸','排氣量','馬力','重量','加速時間','年份']] = arr
df
```

## 最大最小值標準化：MinMaxScaler

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv('汽車車型資料檔.csv')
std = MinMaxScaler()
df1 = df[['哩/加侖','氣缸','排氣量','馬力','重量','加速時間','年份']]
arr = std.fit_transform(df1)
df[['哩/加侖','氣缸','排氣量','馬力','重量','加速時間','年份']] = arr
df
```

## 最大絕對值標準化：MaxAbsScaler

```python
import pandas as pd
from sklearn.preprocessing import MaxAbsScaler

df = pd.read_csv('汽車車型資料檔.csv')
std = MaxAbsScaler()
df1 = df[['哩/加侖','氣缸','排氣量','馬力','重量','加速時間','年份']]
arr = std.fit_transform(df1)
df[['哩/加侖','氣缸','排氣量','馬力','重量','加速時間','年份']] = arr
df
```

## RobustScaler 標準化：RobustScaler

```python
import pandas as pd
from sklearn.preprocessing import RobustScaler

df = pd.read_csv('汽車車型資料檔.csv')
std = RobustScaler()
df1 = df[['哩/加侖','氣缸','排氣量','馬力','重量','加速時間','年份']]
arr = std.fit_transform(df1)
df[['哩/加侖','氣缸','排氣量','馬力','重量','加速時間','年份']] = arr
df
```

## 對應字典法

```python
import pandas as pd
df = pd.read_csv('客戶聯絡狀況資料檔.csv')

df['婚姻'].unique()
```

```python
dict1 = {'單身':1, '已婚':2, '離婚':3, '未知':4}
df['婚姻'].replace(dict1, inplace=True)
df
```

## 標籤編碼法

```python
import pandas as pd
df = pd.read_csv('客戶聯絡狀況資料檔.csv')

df['工作'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
label1 = LabelEncoder()
df['工作'] = label1.fit_transform(df['工作'])
df
```

```python
label1.classes_
```

```python
df['工作'] = label1.inverse_transform(df['工作'])
df
```

```python
import numpy as np
cols = df.select_dtypes(exclude=[np.number]).columns
cols
```

```python
for i in cols:
  df[i] = label1.fit_transform(df[i])
  print('「{}」特徵對照：'.format(i), end='')
  for j in range(len(label1.classes_)):
    print('{} - {}'.format(j, label1.classes_[j]), end=', ')
  print()
df
```

## One-Hot 編碼法

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

df = pd.read_csv('客戶聯絡狀況資料檔.csv')
label1 = LabelEncoder()
df['訂購'] = label1.fit_transform(df['訂購'])
onehot = OneHotEncoder(sparse=False)
arr = onehot.fit_transform(df[['訂購']])
arr
```

```python
df1 = pd.DataFrame(arr, columns=onehot.get_feature_names_out(['訂購']))
df1
```

```python
df.drop('訂購', axis=1, inplace=True)
df2 = pd.merge(df, df1, left_index=True, right_index=True)
df2
```

# 使用Pandas 進行特徵選擇

## 皮爾森(Pearson) 相關係數

```python
import pandas as pd
df = pd.read_csv('地區房價資料檔.csv')
featuresCorr = df.corr()
featuresCorr
```

```python
targetCorr = featuresCorr['房價']
targetCorr = targetCorr.drop('房價')
selectedFeatures = targetCorr[abs(targetCorr) > 0.4]
print("選擇特徵數： {} \n選擇特徵:\n{}".
      format(len(selectedFeatures), selectedFeatures))
```

# 使用Scikit-Learn 進行特徵選擇

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
df = pd.read_csv('乳癌病歷資料檔.csv')
data = df.iloc[:, :-1]
data
```

```python
target = df.iloc[:, -1]
target
```

```python
n = 5
chi = SelectKBest(chi2, k=n)
arrchi = chi.fit_transform(data, target)
arrchi
```

```python
score = chi.scores_
score
```

```python
scoresort = np.argsort(score)
scoresort = np.flipud(scoresort)
scoresort
```

```python
col = df.columns
print('選擇的特徵：')
for i in range(n):
  print('{}：{}'.format(col[scoresort[i]], score[scoresort[i]]))
```
