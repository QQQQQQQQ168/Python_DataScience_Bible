## 長條圖與橫條圖

```python
import matplotlib
import matplotlib.pyplot as plt 
from matplotlib.font_manager import fontManager
fontManager.addfont('NotoSansTC-Regular.otf')
matplotlib.rc('font', family='Noto Sans TC')

listx = ['c','c++','c#','java','python']
listy = [45,28,38,32,50]
plt.bar(listx, listy, width=0.5, color=['r','g','b'])
plt.title("資訊程式課程選修人數")
plt.xlabel("程式課程")
plt.ylabel("選修人數")
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt 
from matplotlib.font_manager import fontManager
fontManager.addfont('NotoSansTC-Regular.otf')
matplotlib.rc('font', family='Noto Sans TC')

listy = ['c','c++','c#','java','python']
listx = [45,28,38,32,50]
plt.barh(listy, listx, height=0.5, color=['r','g','b'])
plt.title("資訊程式課程選修人數")
plt.xlabel("程式課程")
plt.ylabel("選修人數")
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt 
from matplotlib.font_manager import fontManager
fontManager.addfont('NotoSansTC-Regular.otf')
matplotlib.rc('font', family='Noto Sans TC')

listx = ['c','c++','c#','java','python']
listy1 = [25,20,20,16,28]
listy2 = [20,8,18,16,22]
plt.bar(listx, listy1, width=0.5, label='男')
plt.bar(listx, listy2, width=0.5, bottom=listy1, label='女')
plt.legend()
plt.title("資訊程式課程選修人數")
plt.xlabel("程式課程")
plt.ylabel("選修人數")
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt 
from matplotlib.font_manager import fontManager
fontManager.addfont('NotoSansTC-Regular.otf')
matplotlib.rc('font', family='Noto Sans TC')

width = 0.25
listx = ['c','c++','c#','java','python']
listx1 = [x - width/2 for x in range(len(listx))]
listx2 = [x + width/2 for x in range(len(listx))]
listy1 = [25,20,20,16,28]
listy2 = [20,8,18,16,22]
plt.bar(listx1, listy1, width, label='男')
plt.bar(listx2, listy2, width, label='女')
plt.xticks(range(len(listx)), labels=listx)
plt.legend()
plt.show()
```

## 圓形圖

```python
import matplotlib
import matplotlib.pyplot as plt 
from matplotlib.font_manager import fontManager
fontManager.addfont('NotoSansTC-Regular.otf')
matplotlib.rc('font', family='Noto Sans TC')

sizes = [25, 30, 15, 10]
labels = ["北部", "西部", "南部", "東部"]
colors = ["red", "green", "blue", "yellow"]
explode = (0, 0, 0.2, 0)
plt.pie(sizes, 
	explode = explode, 
	labels = labels, 
	colors = colors,
	labeldistance = 1.1, 
	autopct = "%2.1f%%", 
	pctdistance = 0.6,
	shadow = True,
	startangle = 90)
plt.show()
```

## 長條圖與橫條圖

```python
import seaborn as sns
sns.set_theme()
tips = sns.load_dataset("tips")
#計算不同天的小費總額
tips_total = tips.groupby("day", as_index=False).sum()
tips_total.head()
```

```python
sns.barplot(data=tips_total, x='day', y='total_bill')
```

```python
sns.barplot(data=tips_total, x='tip', y='day')
```
